## **The Relation of Diabetes status with different age groups**

For individuals with pre-diabetes (Diabetes_Status = 1), the mean age group is approximately 9.08 that is 60 to 64 years old, and the median age group is 9.
For individuals with diabetes (Diabetes_Status = 2), the mean age group is approximately 9.38 that is 60 to 66 years old, and the median age group is is 10 that is 65 to 69 years old.
This suggests that, on average, individuals with pre-diabetes and diabetes tend to be slightly older compared to those without diabetes. Additionally, the median age gives a measure of the central tendency of age within each group, indicating the age that separates the higher and lower halves of the distribution.

## **Relation of Diabetes status and Blood pressure**

```python
high_bp_counts = df.groupby('Diabetes_Status')['High_Blood_Pressure'].value_counts()
high_bp_counts
```

```python
#Counting occurrences of each combination of 'High BP' and 'Diabetes_status'
grouped_counts = df.groupby(['High_Blood_Pressure', 'Diabetes_Status']).size().unstack()

#Plotting the pie charts for each combination
BP_labels = {0: 'Low', 1: 'High'}
diabetes_labels = ['Non-Diabetic (0)', 'Prediabetic (1)', 'Diabetic (2)']

for i in grouped_counts.index:
    counts = grouped_counts.loc[i]
    plt.figure(figsize=(8, 6))
    plt.pie(counts, labels=diabetes_labels, autopct='%1.1f%%', startangle=90)
    plt.title(f'Diabetes Status Distribution with blood pressure: {BP_labels[i]}')
    plt.axis('equal')
    plt.show()
```

The output indicates a significant association between diabetes and high blood pressure, with diabetic individuals showing a notably higher prevalence of high blood pressure, especially among those with more severe diabetes.

## **Relation of diabetes with being overweight**

```python
#Filtering DataFrame to include individuals aged between 1 and 9 age groups with BMI >= 25
filtered_df = df[(df['Age'] >= 1) & (df['Age'] <= 9) & (df['Body_Mass_Index'] >= 25)]

#Counting occurrences of each Diabetes_Status within the filtered subset
bmi_diabetes_counts = filtered_df['Diabetes_Status'].value_counts()
bmi_diabetes_counts
```

```python
#Calculating the percentages
bmi_diabetes_percentages = filtered_df['Diabetes_Status'].value_counts(normalize=True) * 100
bmi_diabetes_percentages
```

```python
#Filtering DataFrame to include individuals aged between 1 and 9 age group with BMI < 25
filtered_df = df[(df['Age'] >= 1) & (df['Age'] <= 9) & (df['Body_Mass_Index'] < 25)]

#Counting occurrences of each Diabetes_Status within the filtered subset and normalize
bmi_diabetes_percentages = filtered_df['Diabetes_Status'].value_counts(normalize=True) * 100
bmi_diabetes_percentages
```

In individuals aged between 18-64 years old, being categorized as overweight, indicated by a BMI greater than or equal to 25, seems to correlate with an elevated likelihood of diabetes or pre-diabetes. Approximately 13.12% of individuals in this group fall into the diabetic category, with an additional 1.79% categorized as pre-diabetic. Conversely, those classified as not overweight, with a BMI less than 25, exhibit notably lower diabetes prevalence, with only 3.26% categorized as diabetic and 0.68% as pre-diabetic.
